#  Our 'Big' Question


<b> Yelp Dataset Challenge Question </b>

If Yelpers write both reviews and tips for a business, which do they write first (what is the distribution as to what comes first).

<b> Why it is important to answer this question or how does this benefit Yelp Business </b>

Its very important for Yelp as company to improve their customer interaction, sales strategy and grow their communities.

Tip has very less information contained within it, Online customers always want to know more information about a business before they visit. Reviews contain lot of information compared to Tip. Details contain information about attributes of business can be well mentioned in a Review compared to a Tip. So knowing what comes first either Tip or Review would benefit Yelp.

# Few Data Issues encountered and fixing those issues in the query
 
<b> Data Comparision using UNIX_TIMESTAMP </b> 
 
 Using a case statement and Comparison of tip date and review date and coming up with a pattern as which comes first.


<b> Multiple Tip Dates Issue </b>

 Multiple rows problem due to many TIP dates, since review can be given only once and tip can be given any number of times, due to this there are     multiple rows for single user_id

 Using row_num() and partition by function to solve the data issue when there is multiple tips for a business given by a user. Hence considering only the oldest tip date. 
 
 
 <b> Code analysis </b>
 
 Using business table as a driving table and joining other tables using ' left join'.
 Displaying required columns for visualization in Tableau.



```python
df_results=spark.sql("select * from (select distinct b.business_id, b1.category, t.business_id as review_business_id, b1.state, r.user_id, r.date as review_date, row_number() over (partition by r.user_id, b.business_id order by unix_timestamp(t.date, 'yyyy-mm-dd') asc) as row_num, unix_timestamp(r.date,'yyyy-mm-dd') as review_dates, t.date as tip_date, unix_timestamp(t.date, 'yyyy-mm-dd') as tip_dates, case when (unix_timestamp(r.date, 'yyyy-mm-dd') <unix_timestamp(t.date, 'yyyy-mm-dd')) then 'Review_First' when(unix_timestamp(r.date, 'yyyy-mm-dd')>unix_timestamp(t.date, 'yyyy-mm-dd')) then 'Tip_First' else 'Unknown' end as result from business b left join review r on r.business_id=b.business_id left join business_1 b1 on b1.business_id=b.business_id left join tip t on t.business_id=r.business_id where r.user_id=t.user_id) as SUB where SUB.row_num=1 ")
```
